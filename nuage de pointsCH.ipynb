```python
    import matplotlib.pyplot as plt
```

```python
# Données fictives
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
```

```python
# Création du nuage de points
plt.scatter(x, y, color='blue', marker='o')
plt.title("Nuage de points")
plt.xlabel("Axe X")
plt.ylabel("Axe Y")
plt.grid(True)
plt.savefig("nuage_de_points.png")  # Sauvegarde l'image
plt.show()
```
